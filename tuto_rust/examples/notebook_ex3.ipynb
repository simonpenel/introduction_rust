# Structures

We will define a structure to describe a day

```rust
#[derive(Debug)]
pub enum Mois {
    Octobre,
    Novembre,
    Decembre,
    Juillet,
    Aout,
}

#[derive(Debug)]
pub struct Jour {
    /// jour de la semaine
    jour_sem: JourSemaine,
    /// jour du mois
    jour_mois: u32,
    /// mois
    mois: Mois,
}

let ajd = Jour {
    jour_sem: JourSemaine::Vendredi,
    jour_mois: 13,
    mois: Mois::Octobre,
};
println!("Aujourd'hui c'est {:?}", ajd);
println!(
        "Aujourd'hui c'est {:?}, le {} du mois d'{:?}",
        ajd.jour_sem, ajd.jour_mois, ajd.mois
    );
```
